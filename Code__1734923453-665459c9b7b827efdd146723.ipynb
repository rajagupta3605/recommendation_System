## 2. Preprocess Clickstream Data

#### Prepare the user clickstream data for the profiler bot.

```python
# Split impressions into clicked and not clicked articles
def process_impressions(row):
    impressions = row.split()
    clicked = [imp.split("-")[0] for imp in impressions if imp.split("-")[1] == "1"]
    return clicked

# Process user click history and impressions
behaviors['ClickedArticles'] = behaviors['Impressions'].apply(process_impressions)
behaviors['HistoryList'] = behaviors['History'].str.split(" ")

# Drop unnecessary columns
behaviors = behaviors.drop(columns=['Impressions', 'History'])

# Display processed data
print("Processed Behaviors Data: \n", behaviors.head())
```

## 3. Article Recommender Bot

#### Recommend articles based on categories or user profiles.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Combine Title and Abstract to create article content
news['Content'] = news['Title'] + " " + news['Abstract']

# Compute TF-IDF vectors for article content
vectorizer = TfidfVectorizer(stop_words="english")
tfidf_matrix = vectorizer.fit_transform(news['Content'])


# Recommend articles based on cosine similarity or handle cold start
def recommend_articles(user_history, top_n=10):
    if not user_history:  # Cold start: No user history available
        return news.sample(n=top_n)[['NewsID', 'Title', 'Category', 'Abstract']]  # TOP N recommendation
    
    # Find indices of articles in user history
    history_indices = [news[news['NewsID'] == nid].index[0] for nid in user_history if nid in news['NewsID'].values]
    if not history_indices:  # Handle case where history doesn't match any articles
        return news.sample(n=top_n)[['NewsID', 'Title', 'Category', 'Abstract']]
    
    # Calculate similarities
    history_tfidf = tfidf_matrix[history_indices]
    similarities = cosine_similarity(history_tfidf, tfidf_matrix)
    scores = similarities.sum(axis=0)
    top_indices = scores.argsort()[-top_n:][::-1]
    return news.iloc[top_indices][['NewsID', 'Title', 'Category', 'Abstract']]

# Example: Recommend for a user
user_history_example = []  # Replace with actual user history ["N39081", "N114581"] OR set user_history_example = []  when No user profile is available
recommendations = recommend_articles(user_history_example)
print("Recommendations: ", recommendations)
```

## 4. User Profiler Bot

#### Update user preferences based on clickstream.

```python
import tkinter as tk
from time import time

# Initialize Clickstream DataFrame
clickstream_data = pd.DataFrame(columns=[
    'UserId', 'SessionID', 'ArticleID', 'Served Rank', 'Click', 'Time Spent (seconds)'
])

# Interaction Tracker Variables
current_user = 1
current_session = 1
active_article = None
start_time = None
accessed_articles = []  # Track the sequence of accessed articles

# Function to start tracking a user's interaction with an article
def start_article_tracking(article_id):
    """Start tracking a user's interaction with an article."""
    global active_article, start_time
    if active_article:
        print(f"Already tracking Article ID {active_article}. Please stop tracking first.")
        return

    # Set the article being tracked and start the timer
    active_article = article_id
    start_time = time()

    print(f"Started tracking Article ID {article_id}")

# Function to stop tracking interaction and update the clickstream data
def stop_article_tracking(article_id, clicked):
    """Stop tracking interaction and update clickstream data."""
    global active_article, start_time, clickstream_data

    if not active_article:
        print("No article is currently being tracked.")
        return

    # Calculate the time spent if the article was clicked
    elapsed_time = time() - start_time if clicked else 0

    # Assign served rank based on the sequence of accessed articles
    served_rank = len(accessed_articles) + 1  # Rank is based on order of access

    # Add the new interaction to the clickstream data
    clickstream_data = pd.concat([
        clickstream_data,
        pd.DataFrame([{
            'UserId': current_user,
            'SessionID': current_session,
            'ArticleID': article_id,
            'Served Rank': served_rank,  # Assign sequential rank based on access order
            'Click': 'Yes' if clicked else 'No',
            'Time Spent (seconds)': elapsed_time
        }])
    ], ignore_index=True)

    # Add the article to the list of accessed articles
    accessed_articles.append(article_id)

    print(f"Stopped tracking Article ID {article_id}. Clicked: {clicked}, Time Spent: {elapsed_time:.2f}s")
    active_article = None
    start_time = None

# Function to analyze clickstream data and extract user interests
def analyze_clickstream():
    """Analyze clickstream data and extract user interests."""
    global clickstream_data

    if clickstream_data.empty:
        print("No interactions to analyze.")
        return

    # Assuming 'news' DataFrame is predefined elsewhere, merge it with clickstream_data
    enriched_data = clickstream_data.merge(news, left_on='ArticleID', right_on='NewsID', how='left')
    category_summary = (
        enriched_data.groupby('Category')['Time Spent (seconds)']
        .sum()
        .sort_values(ascending=False)
        .reset_index()
    )

    print("\nUser Interests (Top Categories):")
    print(category_summary)

    # Save clickstream data for further use
    clickstream_data.to_csv("clickstream_data.csv", index=False)
    print("\nClickstream data saved to 'clickstream_data.csv'")

    return category_summary

# Function to create the GUI for interacting with articles and recording user clickstream
def create_gui(user_history_example):
    """Create a GUI for interacting with articles and recording user clickstream."""
    global news, current_user, current_session

    # Create the main Tkinter window first
    root = tk.Tk()
    root.title("User Interaction Tracker")

    # Create a list of users for selection (User IDs from 1 to 5)
    users = [1, 2, 3, 4, 5]
    user_var = tk.IntVar(value=1)  # Default to User 1

    # Handler for changing the user selection
    def user_changed(*args):
        """Handler to update the current user and reset session."""
        global current_user, current_session, accessed_articles
        current_user = user_var.get()
        current_session = 1  # Reset session for each user
        accessed_articles = []  # Reset accessed articles for each user

    user_menu = tk.OptionMenu(root, user_var, *users)
    user_menu.pack(pady=10)
    user_var.trace_add("write", user_changed)

    # Function to increment session ID for the current user
    def increment_session():
        """Increment session ID for the current user."""
        global current_session
        current_session += 1

    # Simulated function to recommend articles based on user history
    recommendations = recommend_articles(user_history_example)

    # Display instructions
    tk.Label(root, text="Click on an article to interact.").pack(pady=10)

    # Display the recommended articles and serve them with sequential ranks (1 to 10)
    for idx, row in recommendations.iterrows():
        article_id = row['NewsID']
        article_title = row['Title']

        # Create a button for each article that tracks interactions
        button = tk.Button(
            root, text=f"{article_title} (ID: {article_id})", fg="blue",
            command=lambda aid=article_id: start_article_tracking(aid)
        )
        button.pack(pady=5)

    # Buttons to stop tracking with or without a click
    tk.Button(root, text="Stop Tracking (No Click)", command=lambda: stop_article_tracking(active_article, clicked=False)).pack(pady=10)
    tk.Button(root, text="Stop Tracking (Clicked)", command=lambda: stop_article_tracking(active_article, clicked=True)).pack(pady=10)
    
    # Button to analyze the clickstream data
    tk.Button(root, text="Analyze Data", command=analyze_clickstream).pack(pady=10)

    # Function to save the data and exit the application
    def exit_application():
        clickstream_data.to_csv("clickstream_data.csv", index=False)
        print("Clickstream data saved to 'clickstream_data.csv'")
        root.destroy()

    tk.Button(root, text="Exit", command=exit_application).pack(pady=10)

    # Start the GUI main loop
    root.mainloop()

# Example Launch: Cold Start User History (Use actual data in place of [])
create_gui(user_history_example=[])
```

## Key Points Summary of the Code

#### Data Loading and Preprocessing:
#### - News articles (news.tsv) and user behavior (behaviors.tsv) datasets are loaded and processed.
#### - Splits the Impressions column into clicked articles (ClickedArticles), and processes History into a list (HistoryList) for easier access.
#### - Merges news and behaviors DataFrames for further analysis.
####
#### TF-IDF Vectorization and Article Recommendation:
#### - Combines article titles and abstracts to create textual features.
#### - Converts them into numerical vectors using Tfidf Vectorizer.
#### - Defines a recommendation function that suggests articles to users based on their history using cosine similarity between article vectors.
#### - If no user history is available (cold start), it uses TOP N Recommendation technique.
####
#### Clickstream Data Collection:
#### - Initializes a clickstream_data DataFrame to track user interactions.
#### - Tracks article interactions by starting and stopping article tracking with timestamps and calculates time spent on each article.
#### - Assigns a "Served Rank" to each article interaction based on the order in which articles are accessed by the user.
####
#### Clickstream Analysis:
#### - Analyzes the clickstream data by merging it with the news dataset to identify user interests (e.g., categories of articles with the most time spent).
#### - Saves the clickstream data to a CSV file (clickstream_data.csv) for further use.
####
#### User Interaction via GUI (Tkinter):
#### - A Tkinter GUI is created where users can interact with a list of recommended articles.
#### - Users can start tracking an article, stop tracking (with or without clicking), and view the total time spent on each article.
#### - Articles are recommended and displayed with sequential ranks (from 1 to 10).
#### - Buttons to analyze clickstream data and save it to a CSV file are included in the GUI.
#### 
#### Functionality for Data Saving and Exiting:
#### - Data can be saved by clicking the "Exit" button in the GUI, which stores the clickstream_data DataFrame to a CSV file.
